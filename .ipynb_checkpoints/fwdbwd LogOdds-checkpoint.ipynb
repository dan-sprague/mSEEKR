# Basepair State Likelihood

$\renewcommand{\vec}[1]{\mathbf{#1}}$

Experiment: Use the forward-backward algorithm to calculate the log-odds for each nucleotide of being in the '+' (query) state. 


## fwd_bwd output 

### Calculate

$$P(x_t|o_{1:t},o_{t+1:T}) \propto P(o_{t+1:T}|x_t) P(x_t|o_{1:t})$$

The forward algorithm calculates 
$$P(x_t | o_{1:t}) = P(o_t|x_t)\sum_{x_{t-1}}P(x_t|x_{t-1})P(x_{t-1},o_{1:t-1})$$

Note: the $P(x_{t-1},o_{1:t-1})$ term is the basis for the recursion in this algorithm.

### Output

Let $\vec{x}$ be a vector of length $L$, where $L$ equal to the length of the transcript being studied. Each $\vec{x}_i$ for $i\in{[0,L]}$ is equal to $\log\frac{p_i}{1-p_i}$ where $p_i$ is the probability of being in '+' state at position $i$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

a = np.loadtxt('./test.txt')
```

```python
sns.set_context('talk',font_scale=1.5)
plt.plot(a,)
plt.axhline(y=0,linestyle='--')
plt.ylabel('Log-Odds')
plt.xlabel('hg kcnq1ot1')
plt.title('hg Repeat E')
```

```python
sns.set_context('talk',font_scale=1.5)
plt.plot(a[a>0],)
plt.axhline(y=0,linestyle='--')
plt.ylabel('Log-Odds')
plt.xlabel('Hits')
plt.title('hg Repeat E, Just Hits')
```
